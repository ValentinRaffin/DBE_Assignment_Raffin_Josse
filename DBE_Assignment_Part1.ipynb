## Tax revenue (from 1910 to 2007)

Taxation is, by and large, the most important source of government revenue in nearly all countries. According to the most recent estimates from the International Centre for Tax and Development, total tax revenues account for more than 80% of total government revenue in about half of the countries in the world â€“ and more than 50% in almost every country.

We will focus on the tax revenue from 1910 to 2007 in: 

- France 
- Sweden 
- United Kingdom
- United States

```python
### Modules installation ###

import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
### We read our CSV file (avaible at : https://ourworldindata.org/taxation)

tax = pd.read_csv("tax-revenue-national-income.csv")

# We check that everighting seems right 
print(tax.head())
print(tax.tail())
```

```python
# We noticed that the dataframe is not excactly how we would like 
# We will seperate it into 4 dataframes they will go from 1910 to 2007

tax_FRA = tax[(tax["Entity"] == "France") & (tax["Year"] >= 1910)]
tax_SWE = tax[(tax["Entity"] == "Sweden") & (tax["Year"] >= 1910)]
tax_UK = tax[(tax["Entity"] == "United Kingdom") & (tax["Year"] >= 1910)]
tax_US = tax[(tax["Entity"] == "United States") & (tax["Year"] >= 1910) & (tax["Year"] < 2008)]

# We check that everything went well 

print(tax_FRA.head())
print(tax_SWE.head())
print(tax_UK.head())
print(tax_US.head())
```

```python
### GRAPH ### 

# Let's plot four different graph of the evolution of tax revenu in the four countries 

mpl.style.use("seaborn") # We are using the seaborn style 

fig = plt.subplots(figsize = (9, 6))
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3)
ax4 = plt.subplot(2, 2, 4)

ax1.plot(tax_FRA["Year"], tax_FRA["Tax Revenue (Piketty (2014))"], color = "blue")
ax2.plot(tax_SWE["Year"], tax_SWE["Tax Revenue (Piketty (2014))"], color = "blue")
ax3.plot(tax_UK["Year"], tax_UK["Tax Revenue (Piketty (2014))"], color = "blue")
ax4.plot(tax_US["Year"], tax_US["Tax Revenue (Piketty (2014))"], color = "blue")

axes = [ax1, ax2, ax3, ax4]
countries = ["France", "Sweden", "United Kingdom", "United States"]

for ax, c in zip(axes, countries):
    ax.set_title(f"Tax revenue in {c} from 1910 to 2007", size = 13, weight = "bold")
    ax.set_yticklabels([])
    ax.grid(visible = False)
    ax.set_yticks([10, 20, 30, 40, 50, 60])
    ax.set_yticklabels(["10%", "20%", "30%", "40%", "50%", "60%"])
    ax.grid(visible = True)
    ax.set_xlabel("Year")
    ax.set_ylabel("Tax revenue")
    
plt.tight_layout()
plt.show() 
```

```python
# Now we will plot the four lines on the same graph 
# However we will focus over a new period: post WWII 
# Indeed, in the afternath of WWII, it seems that tax revenues had dramatically
# A focus over this period will be more accurate for a comparaison 

tax_FRA = tax_FRA[tax_FRA["Year"] >= 1950]
tax_SWE = tax_SWE[tax_SWE["Year"] >= 1950]
tax_UK = tax_UK[tax_UK["Year"] >= 1950]
tax_US = tax_US[tax_US["Year"] >= 1950]

# We check that everything went well 

print(tax_FRA.head())
print(tax_SWE.head())
print(tax_UK.head())
print(tax_US.head())
```

```python
### COMPARAISON GRAPH ###

fig, ax = plt.subplots(figsize = (9,6))
ax.plot(tax_FRA["Year"], tax_FRA["Tax Revenue (Piketty (2014))"], label = "France", color = "blue", linewidth = 3)
ax.plot(tax_SWE["Year"], tax_SWE["Tax Revenue (Piketty (2014))"], label = "Sweden", color = "yellow", linewidth = 3)
ax.plot(tax_UK["Year"], tax_UK["Tax Revenue (Piketty (2014))"], label = "United Kingdom", color = "black", linewidth = 3)
ax.plot(tax_US["Year"], tax_US["Tax Revenue (Piketty (2014))"], label = "United States", color = "red", linewidth = 3)

ax.set_yticklabels([])
ax.grid(visible = False)
ax.set_yticks([20, 30, 40, 50, 60])
ax.set_yticklabels(["20%", "30%", "40%", "50%", "60%"])
ax.grid(visible = True)
ax.legend(frameon = True, facecolor = "white", edgecolor = "black", loc = "upper left")
ax.set_xlabel("Year", size = 13)
ax.set_ylabel("Tax revenue", size = 13)
ax.set_title("Tax revenue from 1950 to 2007", size = 16, weight = "bold")

plt.show()
```

```python
# We notice various variations on the U.K. line 
# Let's look at it from a political point of view
# We will show the line in red when the Prime Minister is from the LP (Labour Party)
# We will show the line in blue when the Prime Minister is a Conservative

tax_C1 = tax_UK[(tax_UK["Year"] >= 1951) & (tax_UK["Year"] <= 1964)]
tax_LP1 = tax_UK[(tax_UK["Year"] >= 1964) & (tax_UK["Year"] <= 1971)]
tax_C2 = tax_UK[(tax_UK["Year"] >= 1971) & (tax_UK["Year"] <= 1974)]
tax_LP2 = tax_UK[(tax_UK["Year"] >= 1974) & (tax_UK["Year"] <= 1979)]
tax_C3 = tax_UK[(tax_UK["Year"] >= 1979) & (tax_UK["Year"] <= 1997)]
tax_LP3 = tax_UK[tax_UK["Year"] >= 1997]

# We plot our lines 

fig, ax = plt.subplots(figsize = (8,5))
ax.plot(tax_LP1["Year"], tax_LP1["Tax Revenue (Piketty (2014))"], color = "red", linewidth = 3, label = "Labour Party")
ax.plot(tax_LP2["Year"], tax_LP2["Tax Revenue (Piketty (2014))"], color = "red", linewidth = 3)
ax.plot(tax_LP3["Year"], tax_LP3["Tax Revenue (Piketty (2014))"], color = "red", linewidth = 3)
ax.plot(tax_C1["Year"], tax_C1["Tax Revenue (Piketty (2014))"], color = "blue", linewidth = 3, label = "Conservative")
ax.plot(tax_C2["Year"], tax_C2["Tax Revenue (Piketty (2014))"], color = "blue", linewidth = 3)
ax.plot(tax_C3["Year"], tax_C3["Tax Revenue (Piketty (2014))"], color = "blue", linewidth = 3)

# We set our x-axis and y-axis

ax.set_yticklabels([])
ax.grid(visible = False)
ax.set_yticks([27.5, 30, 32.5, 35, 37.5, 40])
ax.set_yticklabels(["27.5%", "30.0%", "32.5%", "35.0%", "37.5%", "40.0%"])
ax.grid(visible = True)
ax.set_xlabel("Year", size = 13)
ax.set_ylabel("Tax revenue", size = 13)

# We will highlight the periods of some famous Prime Minister 

ax.axvspan(xmin = 1951, xmax = 1955, ymin = 0, alpha = 0.1, color = "blue") # Churchill 
ax.text(1948.5, 28.75, "CHURCHILL", size = 14, weight = "bold", color = "blue")
ax.axvspan(xmin = 1979, xmax = 1990, ymin = 0, alpha = 0.1, color = "blue") # Thatcher
ax.text(1979.75, 28.75, "THATCHER", size = 14, weight = "bold", color = "blue")
ax.axvspan(xmin = 1997, xmax = 2007, ymin = 0, alpha = 0.1, color = "red") # Blair
ax.text(1999.25, 28.75, "BLAIR", size = 14, weight = "bold", color = "red")

# We add the legend and the title 

ax.legend(frameon = True, facecolor = "white", edgecolor = "black")
ax.set_title("Tax revenue from 1951 to 2007 in the U.K.", size = 16, weight = "bold")

# We add a signature 

ax.text(1942, 25.25, "RAFFIN" + " "*134 + "Source: Piketty (2014)", color = "white", backgroundcolor = "black", size=11)

# We plot the graph

plt.show()
```

```python
### We compare the variation of tax revenue in the U.K. ###

# We create a function that calculates the variation (in %) between the beginning period and ending period

def variation(df):
    delta = df["Tax Revenue (Piketty (2014))"].iloc[-1] - df["Tax Revenue (Piketty (2014))"].iloc[1]
    x = delta / df["Tax Revenue (Piketty (2014))"].iloc[1]
    return round(x*100, 2)

# We create a list of the variations in tax revenues when the PM was from the LP 

list_lp = []
list_lp.append(variation(tax_LP1))
list_lp.append(variation(tax_LP2))
list_lp.append(variation(tax_LP3))

# We create a list of the variations in tax revenues when the PM was from a Conservative 

list_cn = []
list_cn.append(variation(tax_C1))
list_cn.append(variation(tax_C2))
list_cn.append(variation(tax_C3))

# We will plot two boxplots to compare the variations in tax revenues between LP and Conservatives 

boxplot_df = pd.DataFrame([[list_cn[0], list_lp[0]], [list_cn[1], list_lp[1]], [list_cn[2], list_lp[2]]], columns = ["Conservatives", "Labour Party"])
box = plt.boxplot(boxplot_df, patch_artist = True)

# We will fill these two boxplots with colors and modify the color/style of the median line 

colors = ["blue", "red"]
for patch, color in zip(box["boxes"], colors):
    patch.set_facecolor(color)
for median in box["medians"]:
    median.set(color = "black", ls = "--")

plt.yticks([-15, -10, -5, 0, 5, 10, 15], ["-15.0%", "-10.0%", "-5.0%", "0.0%", "5.0%", "10.0%", "15.0%"])
plt.ylabel("Tax revenues variations", size = 14, weight = "bold")
plt.xticks([1, 2], ["Conservatives", "Labour Party"], size = 13)
plt.xlabel("Party", size = 14, weight = "bold")
plt.title("Comparaison of variations in tax revenues between Conservatives and LP \n (from 1951 to 2007)", size = 15, weight = "bold")

plt.axhline(color = "grey", alpha = 0.5, lw = 1.5, ls = "--")
plt.show()
```
